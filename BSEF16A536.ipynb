```python
#Improt the required Libraries
import os
import numpy as np
from collections import Counter as ct
```

```python
#Reading the directory of our dataset of documents
files = os.listdir('data/')
files
```

```python
# The following code takes all files, replaces new line character with space character and put words of all files in the Words list 
Words = []
for f in files:
    fhand = open("data/"+f)
    file = fhand.read()
    file = file.replace("\n"," ")
    for word in file.split(" "):
        Words.append(word)

Words[:100]
```

```python
len(Words)
```

```python
#find count of unique words in all these files and store it in a dictionary
unique_words = np.unique(Words)
freq=0

Word_To_Count = {}
for w in Words:
    if w not in Word_To_Count:
        Word_To_Count[w] = 1   #Copy to Dictionary if not available.
    else:
        freq = Word_To_Count[w]
        Word_To_Count[w] = freq+1   #Increase its Frequency by 1 if its available
        
Word_To_Count
```

```python
#print the count of Sweden
Word_To_Count["Sweden"]
```

```python
#convert all text into lower case then find count of unique words,
Small_Words = []
#read each file, split all words and append the lower case word in Small_Words list

for f in files:
    fhand = open("data/"+f)
    file = fhand.read()
    file = file.replace("\n"," ")
    for word in file.split(" "):
        word = word.lower()     #converting each word to LowerCase
        Small_Words.append(word)
```

```python
print(Small_Words[:200])  #Showing first 200 words in lowercase
```

```python
#find count of unique words after converting them to lowercase in all these files and store it in a dictionary

small_unique_words = np.unique(Small_Words)
freq=0

Small_Word_To_Count = {}
for w in Small_Words:
    if w not in Small_Word_To_Count:
        Small_Word_To_Count[w] = 1
    else:
        freq = Small_Word_To_Count[w]
        Small_Word_To_Count[w] = freq+1

#Word_To_Count
#len(unique_words)
len(Small_Word_To_Count)
```

```python
print(Small_Word_To_Count)
```

```python
#sort the small words (in Small_Word_To_Count dictionary) by frequencies and store the results in freq.txt file
#Each line of the file should have word seperated by comma with its counts
#Hint: Can you use tuples to sort key value pairs in a list?
fhand = open("freq.txt",mode='w')
Sort_By_Frequency = sorted(Small_Word_To_Count.items(), key = lambda x: x[1])
#Sort_By_Frequency
for var in Sort_By_Frequency:
    fhand.write("%s , %s"%(var[0],var[1]) + '\n')
fhand.close()

```

```python
#sort the words by alphabatical order and store results in words.txt file
#Each line of the file should have word seperated by comma with its counts
fhand = open("words.txt",mode='w')
Sort_By_Name = sorted(Small_Word_To_Count.items(), key = lambda x: x[0])

for var in Sort_By_Name:
    fhand.write("%s , %s"%(var[0],var[1]) + '\n')
fhand.close()

```

```python
# Sort the list of Small_Words alphabatically and store it in a new list called Vocabulary
# Hint: Explore Set Function

Vocabulary = sorted(Small_Word_To_Count)
Vocabulary
```
